```python
# load all the necessary libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patheffects as path_effects
```

```python
# Load the Router_data csv into the dataframe
Router_df = pd.read_csv('Router_data.csv')
Router_df.head()
```

```python
sums = Router_df.groupby(Router_df["Product Name;"])["Number Of Bugs"].sum()
plt.axis('equal')
plt.pie(sums, labels=sums.index, autopct='%1.1f%%', shadow=True, startangle=140, radius = 2.0)
plt.show()
```

```python
# Bar chart for Average MPG by cylinders
sums.plot(color='blue', kind='bar')
plt.xlabel('Product Name',)
plt.ylabel('Number of Bugs')
plt.title('Number of Bugs by Product', color='black')
plt.legend()
plt.show()
```

```python
# Line chart on same data
sums.plot(color='red', linewidth=5,  
             path_effects=[path_effects.SimpleLineShadow(),
                       path_effects.Normal()])
plt.xlabel('Product Name')
plt.ylabel('Number of Bugs')
plt.title('Number of Bugs by Product')
plt.legend()
plt.show()
```
