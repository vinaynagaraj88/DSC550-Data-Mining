```python
# Import necessary libraries
import numpy as np
import pandas as pd
```

```python
# 1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). 
# Load it to a DataFrame.
random_array = np.random.randint(1,501, size=(10,50))
```

```python
random_array
```

```python
array_df = pd.DataFrame(random_array)
```

```python
array_df
```

```python
# 2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

#Sum of each row
sum_each_row = array_df.sum(axis=1)
sum_each_row
```

```python
# Sum of each column

sum_each_col = array_df.sum(axis=0)
sum_each_col
```

```python
# Sum of all entries

sum_all_entry = array_df.sum().sum()
sum_all_entry
```

```python
# 3. Determine the minimum; maximum; average; and standard deviation of all entries.

df_min = array_df.min().min()
df_max = array_df.max().max()
df_avg = array_df.mean().mean()
df_std = array_df.std().std()
```

```python
# Minimum value of all entries

df_min
```

```python
# Maximum value of all entries

df_max
```

```python
# Average of all entries

df_avg
```

```python
# Standard deviation of all entries

df_std
```

```python
array_df.stack().std()
```

```python
# 4. Sort the DataFrame: by rows, by columns, on row 2, on column 2.

# By Columns
sort_by_col = pd.DataFrame(np.sort(array_df.values, axis=0), index=array_df.index, columns=array_df.columns)
sort_by_col
```

```python
# By Rows

sort_by_row = pd.DataFrame(np.sort(array_df.values, axis=1), index=array_df.index, columns=array_df.columns)
sort_by_row
```

```python
# By row2

sort_by_row2 = array_df.sort_values(by=2, axis=1)
sort_by_row2
```

```python
# By column 2
sort_by_col2 = array_df.sort_values(by=2, axis=0)
sort_by_col2
```
